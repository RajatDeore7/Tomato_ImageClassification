code to: Display all the images from a folder

steps
1. Access the folder/ path ----os
2. Read the images --- opencv---cv2--imread()
3. display ---matplotlit--- imshow()

```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
#path
path="C:/Users/Dell/OneDrive/Documents/Tomato/Testing"
images=os.listdir(path)
type(images)
len(images)
```

```python
img_data=[]
for img in images:
    img_arr=cv2.imread(os.path.join(path,img))
    img_data.append(img_arr)
    
```

```python
fig = plt.figure(figsize=(150,150))
columns = 1 
for i in range(len(img_data)):
    ax = fig.add_subplot(len(images) / columns + 1, columns, i + 1)
    plt.imshow(img_data[i])
    ax.set_title('axes title')
#     plt.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')
    plt.xticks([])
    plt.yticks([])
```
